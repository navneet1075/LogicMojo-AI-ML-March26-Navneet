Question 1:- A digital payment platform recorded the following transaction amounts made by a customer during a single day.

transactions = [1200, 3000, 250, 7800, 540, 3000, 9100, 250, 1500, 7800]

Write a Python program that performs the following analysis:

Identify and display all transaction values that appear more than once.
Detect transactions greater than ₹5000 as these may require fraud monitoring.
Calculate the total spending for the day.
Create a new list containing only unique transaction values.
Sort the unique transaction list in descending order to view the largest transactions first.
    

```python
transactions = [1200, 3000, 250, 7800, 540, 3000, 9100, 250, 1500, 7800]


duplicates = []
fraud_monitoring = []
unique_transactions = []
total_spending = 0

for amt in transactions:

    total_spending += amt
    if amt > 5000:
        fraud_monitoring.append(amt)
    if amt not in unique_transactions:
        unique_transactions.append(amt)
    elif amt not in duplicates:
        duplicates.append(amt)

unique_transactions.sort(reverse=True)

print(f"Duplicate values: {duplicates}")
print(f"Fraud monitoring (>₹5000): {fraud_monitoring}")
print(f"Total spending: ₹{total_spending}")
print(f"Unique transactions (Sorted): {unique_transactions}")
```

Question 2:- A dataset exported from a database contains the following customer age values.

ages = [25, 42, -1, 34, None, 52, 17, -5, 29, None, 46] Some entries are invalid because they are either negative numbers or missing values (None).

Write a Python program that:

Iterates through the dataset and removes all invalid age values.
Stores only valid ages in a new list.
Calculates the average age of valid customers.
Displays the list of customers whose age is greater than 30.
Prints how many invalid entries were removed during the cleaning process.

```python
ages = [25, 42, -1, 34, None, 52, 17, -5, 29, None, 46]

valid_ages = []
invalid_count = 0

for age in ages:
    if age is None or age < 0:
        invalid_count += 1
    else:
        valid_ages.append(age)

total_age = 0
for age in valid_ages:
    total_age += age

if len(valid_ages) > 0:
    average_age = total_age / len(valid_ages)
else:
    average_age = 0

above_30 = []
for age in valid_ages:
    if age > 30:
        above_30.append(age)

print("Valid Ages:", valid_ages)
print("Average Age:", average_age)
print("Ages > 30:", above_30)
print("Invalid entries removed:", invalid_count)
```

Question 3:- An analyst wants to compare the price of products listed on two e-commerce platforms.

The following data is provided: products = ["Laptop","Phone","Tablet","Monitor","Headphones"] amazon_prices = [75000, 45000, 30000, 20000, 5000] flipkart_prices = [72000, 47000, 31000, 19500, 4800]

Write a Python program that:

Combines the product names and their prices into a single data structure representing a table.
Compares the price of each product across both platforms.
Determines which platform offers the lower price.
Calculates the difference between the two prices.
Displays the product name, cheaper platform, and price difference.

```python
products = ["Laptop", "Phone", "Tablet", "Monitor", "Headphones"]
amazon_prices = [75000, 45000, 30000, 20000, 5000]
flipkart_prices = [72000, 47000, 31000, 19500, 4800]

comparison_table = []

for i in range(len(products)):
    product_data = {
        "name": products[i],
        "amazon": amazon_prices[i],
        "flipkart": flipkart_prices[i]
    }
    comparison_table.append(product_data)

for item in comparison_table:
    name = item["name"]
    a_price = item["amazon"]
    f_price = item["flipkart"]
    
    if a_price < f_price:
        cheaper_platform = "Amazon"
        difference = f_price - a_price
    elif f_price < a_price:
        cheaper_platform = "Flipkart"
        difference = a_price - f_price
    else:
        cheaper_platform = "Both (Same Price)"
        difference = 0
        
    print(f"Product: {name} | Cheaper: {cheaper_platform} | Difference: ₹{difference}")
```

Question 4:- A web platform records the hours when users log into the system.

login_hours = [1, 3, 5, 7, 2, 6, 8, 4, 9, 10] Write a Python program that:

Separates login hours into even-hour logins and odd-hour logins.
Identifies login hours that fall into peak activity time (hours greater than 5).
Sorts the login hours in ascending order.
Calculates how many logins occurred during peak hours.

```python
login_hours = [1, 3, 5, 7, 2, 6, 8, 4, 9, 10]

even_hours = []
odd_hours = []
peak_hours = []
peak_count = 0

for hour in login_hours:
    if hour % 2 == 0:
        even_hours.append(hour)
    else:
        odd_hours.append(hour)

for hour in login_hours:
    if hour > 5:
        peak_hours.append(hour)
        peak_count += 1

sorted_hours = list(login_hours)
for i in range(len(sorted_hours)):
    for j in range(0, len(sorted_hours) - i - 1):
        if sorted_hours[j] > sorted_hours[j + 1]:
            sorted_hours[j], sorted_hours[j + 1] = sorted_hours[j + 1], sorted_hours[j]

print("Even-hour logins:", even_hours)
print("Odd-hour logins:", odd_hours)
print("Peak activity hours (>5):", peak_hours)
print("Total peak hour logins:", peak_count)
print("Sorted login hours:", sorted_hours)
```

Question 5:- An e-commerce company exports its daily order log in the following format. Each order contains multiple pieces of information:

orders = [ ["ORD001","Laptop",75000,"Completed"], ["ORD002","Mouse",500,"Completed"], ["ORD003","Laptop",75000,"Returned"], ["ORD004","Keyboard",1500,"Completed"], ["ORD005","Laptop",75000,"Completed"], ["ORD006","Monitor",20000,"Fraud"], ["ORD007","Mouse",500,"Completed"], ["ORD008","Laptop",75000,"Completed"] ]

Each row represents an order with the following structure:

[Order_ID , Product_Name , Price , Status]

Where: Order_ID → unique order identifier Product_Name → product purchased Price → price of the product Status → order status (Completed, Returned, Fraud)

Your Task:

Write a Python program to simulate a simple order analysis system using loops, nested loops, conditional statements, and list operations.

Your program must perform the following steps.

Display All Orders Using a nested loop, print the entire order table row by row so that each order’s details are displayed clearly.

Detect Suspicious Activity The system suspects fraud when: the same product appears more than 3 times, OR an order has status "Fraud"

Write a program that scans the dataset and identifies suspicious products or orders. You must use nested loops to compare product occurrences.

Calculate Total Revenue Only Completed orders contribute to revenue. Write code that: iterates through the dataset adds the price of all completed orders prints the total revenue generated for the day.

Count Order Types Using loops and conditional statements, count how many orders fall into each category:

Completed Returned Fraud Store these counts in separate variables.

Generate a List of Unique Products Create a new list containing unique product names present in the dataset. You must implement this without using the set() function.

Product Sales Analysis For each unique product, determine: how many times it was ordered total revenue generated by that product This requires nested loops.

Identify the Most Sold Product Based on the frequency of orders, determine which product appears the most in the dataset.

Remove Fraud Orders Create a cleaned dataset by removing orders whose status is "Fraud". Use list methods to store the cleaned data in a new list.

```python
orders = [
    ["ORD001", "Laptop", 75000, "Completed"],
    ["ORD002", "Mouse", 500, "Completed"],
    ["ORD003", "Laptop", 75000, "Returned"],
    ["ORD004", "Keyboard", 1500, "Completed"],
    ["ORD005", "Laptop", 75000, "Completed"],
    ["ORD006", "Monitor", 20000, "Fraud"],
    ["ORD007", "Mouse", 500, "Completed"],
    ["ORD008", "Laptop", 75000, "Completed"]
]

print("--- Order Table ---")
for row in orders:
    for detail in row:
        print(detail, end=" | ")
    print()

suspicious_products = []
for i in range(len(orders)):
    product_name = orders[i][1]
    count = 0
    for j in range(len(orders)):
        if orders[j][1] == product_name:
            count += 1
    if count > 3 and product_name not in suspicious_products:
        suspicious_products.append(product_name)

print("\n--- Suspicious Activity ---")
if suspicious_products:
    print("Products ordered > 3 times:", suspicious_products)

for order in orders:
    if order[3] == "Fraud":
        print(f"Fraud Order Detected: {order[0]} for {order[1]}")

total_revenue = 0
completed_count = 0
returned_count = 0
fraud_count = 0

for order in orders:
    status = order[3]
    price = order[2]
    if status == "Completed":
        total_revenue += price
        completed_count += 1
    elif status == "Returned":
        returned_count += 1
    elif status == "Fraud":
        fraud_count += 1

print("\n--- Revenue and Status Counts ---")
print(f"Total Revenue: ₹{total_revenue}")
print(f"Completed: {completed_count}, Returned: {returned_count}, Fraud: {fraud_count}")

unique_products = []
for order in orders:
    product = order[1]
    if product not in unique_products:
        unique_products.append(product)

print("\n--- Product Sales Analysis ---")
most_sold_product = ""
max_sales_count = 0

for prod in unique_products:
    prod_count = 0
    prod_revenue = 0
    for order in orders:
        if order[1] == prod:
            prod_count += 1
            if order[3] == "Completed":
                prod_revenue += order[2]
    
    print(f"Product: {prod} | Quantity: {prod_count} | Revenue: ₹{prod_revenue}")
    
    if prod_count > max_sales_count:
        max_sales_count = prod_count
        most_sold_product = prod

print(f"\nMost Sold Product: {most_sold_product} ({max_sales_count} times)")

cleaned_orders = []
for order in orders:
    if order[3] != "Fraud":
        cleaned_orders.append(order)

print("\n--- Cleaned Dataset (Fraud Removed) ---")
for row in cleaned_orders:
    print(row)
```
